```python
# 필요한 라이브러리를 임포트합니다.
import qrcode
import matplotlib.pyplot as plt
from google.colab import drive

# 구글 드라이브에 연결합니다.
drive.mount('/content/drive')

# QR코드로 만들고 싶은 데이터를 입력합니다.
data = "https://padlet.com/kcookusb/introduce-yourself-ywtzx5bz0f9wsw2d"

# QR코드 객체를 생성합니다.
qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_L,
    box_size=10,
    border=4,
)

# QR코드 객체에 데이터를 추가합니다.
qr.add_data(data)
qr.make(fit=True)

# QR코드 이미지를 생성합니다.
img = qr.make_image(fill_color="black", back_color="white")

# QR코드 이미지를 구글 드라이브에 저장합니다.
img.save("/content/drive/MyDrive/qr_code.png")

# QR코드 이미지를 colab에서 보여줍니다.
plt.imshow(img, cmap="gray")
plt.axis("off")
plt.show()
```

```python
import gradio as gr
from gtts import gTTS
from pydub import AudioSegment
import os

def text_to_speech(text):
    tts = gTTS(text=text, lang='en')
    mp3_filename = "voice.mp3"
    wav_filename = "voice.wav"
    tts.save(mp3_filename)

    # Convert from MP3 to WAV
    audio = AudioSegment.from_mp3(mp3_filename)
    audio.export(wav_filename, format="wav")

    return wav_filename

iface = gr.Interface(fn=text_to_speech,
                     inputs=gr.inputs.Textbox(),
                     outputs=gr.outputs.Audio(type="filepath"))
iface.launch(share=True)
```
